# Breakdown Analytics Notebook
Notebook ini dibuat untuk workflow mirip Google Colab: load data, cleaning, classification, KPI, dan visualisasi.

## 1. Setup

```python
from pathlib import Path
import pandas as pd
import plotly.io as pio

pio.renderers.default = 'browser'

from scripts.cleaning import load_raw_data, clean_and_transform
from scripts.classify import add_classification_columns
from scripts.analytics import compute_kpis
from scripts.visualization import (
    chart_pareto_category,
    chart_daily_trend,
    chart_top_units,
    chart_by_location,
    chart_heatmap,
    chart_timeline,
)

base_dir = Path.cwd()
raw_dir = base_dir / 'data' / 'raw'
raw_dir
```

## 2. Load Raw Data

```python
df_raw = load_raw_data(raw_dir)
print(f'Rows: {len(df_raw):,}')
print(f'Columns: {len(df_raw.columns)}')
df_raw.head()
```

## 3. Cleaning + Transform

```python
df_clean = clean_and_transform(df_raw)
df_clean[['Date','CN Unit','Awal','Akhir','Hours','Duration_Real','Duration_Check']].head(10)
```

## 4. Classification + Derived Columns

```python
df = add_classification_columns(df_clean)
df[['Description of Breakdown','Category','Severity','Month','Week','Hour_Start','Shift','Shift_Num']].head(15)
```

## 5. Data Quality Checks

```python
display(df['Duration_Check'].value_counts(dropna=False))
display(df['Invalid_Time'].value_counts(dropna=False))
display(df['Category'].value_counts().head(10))
```

## 6. KPI Summary

```python
kpi = compute_kpis(df)
kpi
```

## 7. Visualizations

```python
chart_pareto_category(df).show()
chart_daily_trend(df).show()
chart_top_units(df).show()
chart_by_location(df).show()
chart_heatmap(df).show()
chart_timeline(df).show()
```

## 8. Save Processed Data

```python
clean_path = base_dir / 'data' / 'clean' / 'breakdown_clean.parquet'
master_path = base_dir / 'data' / 'master' / 'breakdown_master.parquet'

df_clean.to_parquet(clean_path, index=False)
df.to_parquet(master_path, index=False)
clean_path, master_path
```
